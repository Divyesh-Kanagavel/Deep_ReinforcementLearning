```python
import gymnasium as gym

env = gym.make("LunarLander-v2") # the Lunar Lander environment

observation, info = env.reset() # Reset the environement to starting state

for _ in range(20):
  action = env.action_space.sample()
  print("Action taken:", action)
  observation, reward, terminated, truncated, info = env.step(action)
  if truncated or terminated:
    observation, info = env.reset()
    print("Environment is reset")
env.close()
```

we kept on taking random actions in the cell above. the objective is to learn a polcy which takes the right action given a state $s_t$.

About the environment :     
The environment is part of the box2D environments. This environment is a classic rocket trajectory optimization problem. According to Pontryagin's maximum principle, it is optimal to fire the engine at full throttle or turn it off. This is the reason why this environment has discrete actions: engine on or off [in lunar lander documentation page.]. \

Action space
There are four discrete actions available:

0: do nothing \
1: fire left orientation engine \
2: fire main engine \
3: fire right orientation engine \

Observation space : \
The state is an 8-dimensional vector: the coordinates of the lander in x & y, its linear velocities in x & y, its angle, its angular velocity, and two booleans that represent whether each leg is in contact with the ground or not. \

Reward:   
For each step, the reward:

Is increased/decreased the closer/further the lander is to the landing pad. \
Is increased/decreased the slower/faster the lander is moving. \
Is decreased the more the lander is tilted (angle not horizontal). \
Is increased by 10 points for each leg that is in contact with the ground. \
Is decreased by 0.03 points each frame a side engine is firing. \
Is decreased by 0.3 points each frame the main engine is firing. \
The episode receive an additional reward of -100 or +100 points for crashing or landing safely respectively. \

An episode is considered a solution if it scores at least 200 points. \

Let us get all these infos from the env in code. \

```python
# We create our environment with gym.make("<name_of_the_environment>")
env = gym.make("LunarLander-v2")
env.reset()
print("_____OBSERVATION SPACE_____ \n")
print("Observation Space Shape", env.observation_space.shape)
print("Sample observation", env.observation_space.sample())  # Get a random observation
```

```python
print("\n _____ACTION SPACE_____ \n")
print("Action Space Shape", env.action_space.n)
print("Action Space Sample", env.action_space.sample())  # Take a random action
```

```python
# Create the environment
env = make_vec_env("LunarLander-v2", n_envs=16) # stacking 16 environments simultaneously to have diverse experiences.
```

To learn the control policy, we are going to use the Stable Baselines 3 library which is a reliable set of RL implementations in pytorch. \
We will be using a SOTA learning algorithm called the PPO (Proximal policy implementation)

```python
# since the input is a vector of values, we use the MLP Policy - multilayer perceptron, if the input had been an image, we could use a CNN Policy.
model = PPO(policy='MlpPolicy', env = env, n_steps = 1024, batch_size=64,
    n_epochs=4,
    gamma=0.999,
    gae_lambda=0.98,
    ent_coef=0.01,
    verbose=1)  # we will get to see what the parameters mean when PPO is discussed in great detail
```

We will be using the GPUs for training

```python
# Train it for 1,000,000 timesteps
model.learn(total_timesteps=1000000) # 1 million timesteps
# Save the model
model_name = "ppo-LunarLander-v2"
model.save(model_name)
```

```python
eval_env = Monitor(gym.make("LunarLander-v2"))
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic=True)
print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```

with our model, we are getting a reward of 251 which is good given the requirement was to get a score of 200. we may not know how to finetune this model without understanding the hyperparamters. Let us see if genAI is able to get us a better model.

```python
# prompt: Hi, you are an expert In deep reinforcement learning, from the cells above in the notebook, we have used a model with the given configuration : model = PPO(policy='MlpPolicy', env = env, n_steps = 1024, batch_size=64,
#     n_epochs=4,
#     gamma=0.999,
#     gae_lambda=0.98,
#     ent_coef=0.01,
#     verbose=1) and model.learn(total_timesteps=1000000)!!! can you pls fine tune the model and the time steps during training?

# Fine-tune the model and training parameters
model = PPO(policy='MlpPolicy', env=env, n_steps=2048, batch_size=128,
            n_epochs=10,  # Increased epochs for better optimization
            gamma=0.99,  # Slightly reduced gamma
            gae_lambda=0.95,  # Adjusted lambda
            ent_coef=0.005,  # Reduced entropy coefficient
            learning_rate=3e-4, # Added learning rate
            clip_range=0.2, # Added clip range
            verbose=1)

# Increased training time steps for more exploration
model.learn(total_timesteps=2000000)

# Save the fine-tuned model
model_name = "ppo-LunarLander-v2-finetuned"
model.save(model_name)

# Evaluate the fine-tuned model
eval_env = Monitor(gym.make("LunarLander-v2"))
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic=True)
print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```
